# Histograma $\times$ Gráficos de Densidade

O gráfico abaixo é bastante semelhante ao histograma produzido antes. Mas note que no eixo y não temos a frequencia absoluta dos valores, mas sua probabilidade! No fundo o que esses gráficos representam é a mesma coisa pois a probabilidade é a frequencia relativa dos valores (sua frequencia absoluta dividida pelo total).

```python
%%R

hist(mtcars$mpg, main='Histograma de mpg',col='lightblue',prob=T)
```

e podemos ainda criar um gráfico contínuo aproximando a curva para intervalos de valores contínuos.

```python
%%R

densityplot(mtcars$mpg, main='Density de mpg',col='red' )
```
